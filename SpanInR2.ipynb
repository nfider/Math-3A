```julia
# Example 2: linear combinations of u,v,w, INCLUDING COLORED VECTORS
using Plots
plotly()
u=[2,-1]; v=[6,4]; w=[5,0];


a1=rand(-5:5); a2=rand(-3:3); a3=rand(-3:3);
U=u*a1; V=v*a2; W=w*a3;

while abs(U[1]+V[1]+W[1])>20 || abs(U[2]+V[2]+W[2])>20
   a1=rand(-5:5)
    a2=rand(-3:3)
    a3=rand(-3:3)
    U=u*a1 
    V=v*a2
    W=w*a3 
end

xu=[0,U[1]];
yu=[0,U[2]];
plot(xu,yu, title="Example 2a: lin.combinations of u,v,w", label="$a1 u", linecolor="red",lims=[-20,20])

xv=[U[1],U[1]+V[1]];
yv=[U[2],U[2]+V[2]];
plot!(xv,yv, linecolor="purple", label="$a2 v")

xw=[U[1]+V[1],U[1]+V[1]+W[1]];
yw=[U[2]+V[2],U[2]+V[2]+W[2]];
plot!(xw,yw, linecolor="blue", label="$a3 w")
scatter!([U[1]+V[1]+W[1], U[1]+V[1]+W[1]],[U[2]+V[2]+W[2],U[2]+V[2]+W[2]], markercolor="black", label="end point")

```

```julia
# Example 2: linear combinations of u,v,w, FINAL POINTS ONLY
using Plots
plotly()
u=[2,-1]; v=[6,4]; w=[5,0];
NumberOfPoints=500;

a1=rand(-5:5); a2=rand(-5:5); a3=rand(-5:5);
U=u*a1; V=v*a2; W=w*a3;

while abs(U[1]+V[1]+W[1])>20 || abs(U[2]+V[2]+W[2])>20
   a1=rand(-5:5)
    a2=rand(-5:5)
    a3=rand(-5:5)
    U=u*a1 
    V=v*a2
    W=w*a3 
    
end
p=plot(title="Example 2a: Span{u,v,w}", legend=false)
scatter!([U[1]+V[1]+W[1], U[1]+V[1]+W[1]],[U[2]+V[2]+W[2],U[2]+V[2]+W[2]], markercolor="black", label="end point",lims=[-20,20])

if NumberOfPoints>1
    for n = 2:NumberOfPoints
        a1=rand(-5:5); a2=rand(-5:5); a3=rand(-5:5);
        U=u*a1; V=v*a2; W=w*a3;
        while abs(U[1]+V[1]+W[1])>20 || abs(U[2]+V[2]+W[2])>20
            a1=rand(-5:5)
            a2=rand(-5:5)
            a3=rand(-5:5)
            U=u*a1 
            V=v*a2
            W=w*a3 
        end
        scatter!([U[1]+V[1]+W[1], U[1]+V[1]+W[1]],[U[2]+V[2]+W[2],U[2]+V[2]+W[2]], markercolor="black", label="end point")
    end
end
display(p)
```
